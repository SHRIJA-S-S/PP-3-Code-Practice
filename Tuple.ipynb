```python
print("tuple")
tuple=()
print("empty tuple:",tuple)
tuple=(5,6,7,8,9,10)
print("2nd tuple:",tuple)
tuple=("shri",20.5,5)
print("mixed tuple:",tuple)
tuple=("shri",20.5,5,(5,6,7,8,9,10))
print("nested tuple:",tuple)
print(tuple)
print(tuple[0])
print(tuple[1])
print(tuple[:2])
print(tuple[-3])
print(tuple[:])
print(tuple[:-1])
print(tuple[:0])
print("type tuple")
print(type(tuple))
single_tuple=("shri")
print(type(single_tuple))
single_tuple=("shri",)
print(type(single_tuple))
single_tuple="shri",
print(type(single_tuple))
```
